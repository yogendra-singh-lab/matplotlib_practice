## import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style


```python
classes = ['python','dbms','os','datascience','ml']
class1_student = [34,26,6,78,34]
class2_student = [45,56,34,26,78]
```

```python
plt.bar(classes,class1_student)
plt.show()
```

```python
style.use('ggplot')
plt.bar(classes,class1_student,width = 0.3, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.bar(classes,class1_student,width = 0.3, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.title("bar graph of students and subjects",fontsize = 20)
plt.xlabel('subjects',fontsize  = 15)
plt.ylabel('no.of students',fontsize = 15)
plt.bar(classes,class1_student,width = 0.3, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.title("bar graph of students and subjects",fontsize = 20)
plt.xlabel('subjects',fontsize  = 15)
plt.ylabel('no.of students',fontsize = 15)
plt.bar(classes,class2_student,width = 0.5, color = 'green',edgecolor = 'black',)
plt.bar(classes,class1_student,width = 0.3, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
classes_index = np.arange(len(classes))
width = 0.4
plt.title("bar graph of students and subjects",fontsize = 20)
plt.xlabel('subjects',fontsize  = 15)
plt.ylabel('no.of students',fontsize = 15)
plt.bar(classes_index,class2_student,width, color = 'green',edgecolor = 'black',)
plt.bar(classes_index + width,class1_student,width, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
classes_index = np.arange(len(classes))
width = 0.4
plt.xticks(classes_index + width, classes)
plt.title("bar graph of students and subjects",fontsize = 20)
plt.xlabel('subjects',fontsize  = 15)
plt.ylabel('no.of students',fontsize = 15)
plt.bar(classes_index,class2_student,width, color = 'green',edgecolor = 'black',)
plt.bar(classes_index + width,class1_student,width, color = 'pink',edgecolor = 'black',)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
classes_index = np.arange(len(classes))
width = 0.4
plt.xticks(classes_index + width, classes,rotation = 30)
plt.title("bar graph of students and subjects",fontsize = 20)
plt.xlabel('subjects',fontsize  = 15)
plt.ylabel('no.of students',fontsize = 15)
plt.bar(classes_index,class2_student,width, color = 'green',edgecolor = 'black',)
plt.bar(classes_index + width,class1_student,width, color = 'pink',edgecolor = 'black',)
plt.show()
```
