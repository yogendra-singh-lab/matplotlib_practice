# Pie Chart


```python
import matplotlib.pyplot as plt
```

```python
classes = ["python","artificial intelligence","java","deep learning","ml"]
class_students = [45,23,56,84,9]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class_students, labels = classes)
plt.show()
```

```python
classes = ["python","artificial intelligence","java","deep learning","ml"]
class_students = [45,23,56,84,9]
explode = [0,0.2,0.1,0,0.2]

plt.pie(class_students, labels = classes,explode = explode, autopct = "%0.2f%%")
plt.show()

```

```python
classes = ["python","artificial intelligence","java","deep learning","ml"]
class_students = [45,23,56,84,9]
explode = [0,0.2,0.1,0,0.2]
textprops = {"fontsize" : 16}

plt.pie(class_students, labels = classes,explode = explode, autopct = "%0.2f%%",textprops = textprops,radius = 3)
plt.show()

```
